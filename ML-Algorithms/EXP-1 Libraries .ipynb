**Scikit-Learn**

```python
import numpy as np
from sklearn import linear_model
# Smaller dataset
X = np.array(range(-10,10))  # or X=np.array(range(-50,50))
Y = np.array([3*x+4 for x in X])
```

```python
# Reshape
X = X.reshape(-1,1) #or Y.reshape(100,-1)
print(X)
print(X.shape)
```

```python
model = linear_model.LinearRegression()  # Model
model.fit(X,Y) #making the model to fit around the body
```

```python
# Outputs
print("Prediction for 2:", model.predict([[2]]))
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
model.predict([[62],[2],[5]])
```

```python
#Linear Regression for Two Input Variables
X1 = np.array(range(-50, 50))
X2 = np.array(range(50, -50, -1))

X = np.column_stack((X1, X2))
Y = np.array([2*x1 + 3*x2 + 5 for x1, x2 in X])

model = linear_model.LinearRegression()
model.fit(X, Y)
print("Prediction:", model.predict([[10, 20]]))
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
