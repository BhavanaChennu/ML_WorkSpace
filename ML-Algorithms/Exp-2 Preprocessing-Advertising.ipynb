```python
#import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#load dataset from google drive
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv("/content/drive/MyDrive/Datasets/Advertising.csv")
df
```

```python
print("First 2 rows:,df.head(2)")  #check first few rows
print("Shape:,df.shape()") #check shape of dataset
print("No.of columns:,df.columns") #check column names
```

```python
#Drop unneccessary column
df=df.drop(columns=['Unnamed: 0'])
print("After Removing :,df.columns")
print("After Removing :,df.shape")
```

```python
#check for duplicate rows
print("Duplicates:",df.duplicated().sum())
df=df.drop_duplicates()
```

```python
#check missing values
print(df.isna().sum())
df1=df.copy()
```
